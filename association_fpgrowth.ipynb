```python
merged = order_products_prior.merge(products[['product_id', 'product_name', 'department_id']], on='product_id', how='left')

selected_departments = [1, 3, 9, 12, 13, 14, 15, 16]
merged = merged[merged['department_id'].isin(selected_departments)]


top_1000_products = merged['product_name'].value_counts().head(1000).index
filtered_data = merged[merged['product_name'].isin(top_1000_products)]


transactions = filtered_data.groupby('order_id')['product_name'].apply(list).tolist()

```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules
te = TransactionEncoder()
te_array = te.fit_transform(transactions)
basket = pd.DataFrame(te_array, columns=te.columns_)


# Only Top 300 products
top_300_products = basket.sum().sort_values(ascending=False).head(300).index

df_top_300 = basket[top_300_products]

order_sizes = df_top_300.sum(axis=1)

df_top = df_top_300[order_sizes > order_sizes.quantile(0.9)]
```

```python
frequent_itemsets = fpgrowth(df_top, min_support=0.01, use_colnames=True)

frequent_itemsets_1 = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: len(x) > 1)]

```
